```r
library(gdata)

df = read.csv("day.csv")
df
```

```r
library(ggplot2)
my.scatter_1 = ggplot(df, aes(temp, hum, color = yr))
my.scatter_1 = my.scatter_1 + layer( geom = "point",
                                   stat = "identity",
                                   position = "identity",
                                   params = list(na.rm = FALSE))
my.scatter_1

my.scatter_2 = ggplot(df, aes(hum, windspeed, color = yr))
my.scatter_2 = my.scatter_2 + layer( geom = "point",
                                   stat = "identity",
                                   position = "identity",
                                   params = list(na.rm = FALSE))
my.scatter_2

my.scatter_3 = ggplot(df, aes(temp, windspeed, color = yr))
my.scatter_3 = my.scatter_3 + layer( geom = "point",
                                   stat = "identity",
                                   position = "identity",
                                   params = list(na.rm = FALSE))
my.scatter_3
```

```r
my.bar_plot_1 = ggplot(df, aes(x = cnt, fill = factor(yr)))
my.bar_plot_1 = my.bar_plot_1 + layer(
  geom = "bar",
  stat = "bin",
  position = "identity",
  params = list(
    binwidth = 500,
    na.rm = FALSE,
    alpha = 0.5)
    ) + facet_wrap(~factor(yr))
my.bar_plot_1

my.bar_plot_2 = ggplot(df, aes(x = casual, fill = factor(yr)))
my.bar_plot_2 = my.bar_plot_2 + layer(
  geom = "bar",
  stat = "bin",
  position = "identity",
  params = list(
    binwidth = 500,
    na.rm = FALSE,
    alpha = 0.5)
    ) + facet_wrap(~factor(yr))
my.bar_plot_2

my.bar_plot_3 = ggplot(df, aes(x = registered, fill = factor(yr)))
my.bar_plot_3 = my.bar_plot_3 + layer(
  geom = "bar",
  stat = "bin",
  position = "identity",
  params = list(
    binwidth = 500,
    na.rm = FALSE,
    alpha = 0.5)
    ) + facet_wrap(~factor(yr))
my.bar_plot_3
```
